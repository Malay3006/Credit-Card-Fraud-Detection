```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Construct the full path to the creditcard.csv file
dataset_file_path = os.path.join(path, 'creditcard.csv')

# Load the dataset into a pandas DataFrame
df = pd.read_csv(dataset_file_path)

# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
legit.value_counts()
```

```python
legit_sample = legit.sample(n=492)
```

```python
credited_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
credited_df.groupby('Class').mean()
```

```python
X = credited_df.drop(columns='Class', axis=1)
Y = credited_df['Class']
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
ypred = model.predict(X_test)
```

```python
accuracy_score(ypred, Y_test)
```
